```python
# This converts the employee work length into just integers and nothing else bascially remove all extra words after the year has been mentioned
#read_loan_data["emp_length"] = read_loan_data["emp_length_int"] 

read_loan_data["emp_length_int"] = read_loan_data["emp_length"].str.replace('+ years','')
read_loan_data["emp_length_int"] = read_loan_data["emp_length"].str.replace('years','')
read_loan_data["emp_length_int"] = read_loan_data["emp_length"].str.replace('< 1 year',str(0))
```

```python
type(read_loan_data['emp_length_int'][0]) #so before the emp_length was all objects right now it has been coverted to string
#Next we need to convert the str to int 
```

```python
# convert to integer because in previous line it was str 
read_loan_data['emp_length_int'] = pd.to_numeric(read_loan_data["emp_length_int"], errors = 'coerce')
# now if you have invallid values u can use this fucntion called: errors ="coerce" to NaN
# the to_numeric fucntion is used to convert str to Numpy 


type(read_loan_data['emp_length_int'][0]) 
```

```python
read_loan_data["earliest_cr_line_date"] = pd.to_datetime(read_loan_data['earliest_cr_line'],format='%b-%y',errors='coerce')
type(read_loan_data['earliest_cr_line_date'][0])
```

```python
# to calculate overall time has been passed after the earliest credit line 
#creates a series showing how long each borrower has had credit history as of Sept 1 3035
#longer days mean better credit history

pd.to_datetime('2025-09-01') - read_loan_data["earliest_cr_line_date"]
```

```python
# now to convert all of the days which a person had credit into month we can do:
# for all months in a year avg days in a month in 30.44 
#the new columns name for number of months is: earliest_cr_line_date_month


read_loan_data["earliest_cr_line_date_month"]= round((pd.to_datetime('2025-09-01') - read_loan_data["earliest_cr_line_date"]).dt.days/30.44)


read_loan_data["earliest_cr_line_date_month"].describe()
```
